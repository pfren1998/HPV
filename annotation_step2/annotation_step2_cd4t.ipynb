```r
library(Seurat)
library(pheatmap)
library(purrr)
library(dplyr)
library(stringr)
library(harmony)
```

```r
top_genes = function(degs){
    rownames(degs) = NULL
    degs = degs[!str_detect(degs$gene,'^RPL'),]
    degs = degs[!str_detect(degs$gene,'^RPS'),]
    for (cluster in unique(degs$cluster)){
        message(cluster)
        print(degs[degs$cluster == cluster,] %>% arrange(desc(avg_log2FC)) %>% head(n=50))
    } 
}
```

```r
cd4t = readRDS('~/HPV/res/public_rna/seurat_rds/annotated/cd4t.rds')
```

```r
Idents(cd4t) = cd4t$minor
degs = FindAllMarkers(cd4t,logfc.threshold = 0.25,return.thresh = 0.05)
```

```r
top_genes(degs)
```

```r
options(repr.plot.width = 5, repr.plot.height = 4)
DimPlot(cd4t, group.by = 'seurat_clusters', label = TRUE,repel = TRUE)
```

```r
options(repr.plot.width = 5.5, repr.plot.height = 4)
DimPlot(cd4t, group.by = 'minor', label = TRUE,repel = TRUE)
```
