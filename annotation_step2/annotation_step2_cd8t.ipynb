```r
top_genes = function(degs){
    rownames(degs) = NULL
    degs = degs[!str_detect(degs$gene,'^RPL'),]
    degs = degs[!str_detect(degs$gene,'^RPS'),]
    for (cluster in unique(degs$cluster)){
        message(cluster)
        print(degs[degs$cluster == cluster,] %>% arrange(desc(avg_log2FC)) %>% head(n=50))
    } 
}
```

```r
Idents(cd8t) = cd8t$minor
degs = FindAllMarkers(cd8t,logfc.threshold = 0.25,return.thresh = 0.05)
```

```r
top_genes(degs)
```

# curate

```r
library(hdf5r)
library(Seurat)
library(purrr)
library(dplyr)
library(stringr)
library(harmony)
library(DoubletFinder)
library(ggplot2)
library(ggpubr)
library(plyr)
library(pheatmap)
```

```r
cd8t = readRDS('~/HPV/res/public_rna/seurat_rds/annotated/cd8t.rds')
```

```r
cluster3_temp = cd8t[,cd8t$seurat_clusters %in% c(0,8,12)]
```

```r
cluster3 = CreateSeuratObject(counts = cluster3_temp@assays$RNA@counts, project = "epi") %>% 
    NormalizeData(verbose = FALSE) %>%
    FindVariableFeatures(selection.method = "vst", nfeatures = 40) 
VariableFeatures(cluster3) = unique(c(VariableFeatures(cluster3),c('TCF7','IL7R','XCL1','CXCR3','PDCD1','ENTPD1','HAVCR2','LAG3','LAYN','TOX','GZMB','PRF1','GZMA','GNLY','IFNG','PRF1','CXCL13')))
    cluster3 = ScaleData(cluster3, verbose = FALSE) %>% 
    RunPCA(pc.genes = epi@var.genes, npcs = 10, verbose = FALSE)
options(repr.plot.width = 6, repr.plot.height = 3)
ElbowPlot(cluster3,ndims = 10)
```

```r
length(unique(VariableFeatures(cluster3)))
```

```r
cluster3@meta.data = cluster3_temp@meta.data
cluster3 = RunHarmony(cluster3, 'Patient', verbose = FALSE) %>%
RunUMAP(reduction = "harmony", dims = 1:6, verbose = FALSE) %>%
    FindNeighbors(reduction = "harmony", dims = 1:6, verbose = FALSE) %>%
    FindClusters(resolution = 0.6, verbose = FALSE)
```

```r
options(repr.plot.width = 12, repr.plot.height = 4)
VlnPlot(cluster3,features = c("nFeature_RNA", "nCount_RNA"),pt.size = 0)
```

```r
options(repr.plot.width = 4, repr.plot.height = 3.5)
DimPlot(cluster3, group.by = 'seurat_clusters',label = TRUE,label.size = 3)
```

```r
options(repr.plot.width = 21, repr.plot.height = 8.5)
FeaturePlot(cluster3,features = c('TCF7','CCR7','XCL1','CXCR3','PDCD1','ENTPD1','HAVCR2','LAG3','LAYN','TOX','GZMB','IFNG','PRF1','GZMA','GNLY','IFNG','PRF1','CXCL13','TBX21','TNF','GZMK'),ncol = 7)
```

```r
table(cluster3$seurat_clusters)
```

```r
options(repr.plot.width = 7, repr.plot.height = 5)
DotPlot(cluster3,features = c('TCF7','IL7R','CXCR3','ENTPD1','GZMB','TOX'))
```

```r
Idents(cluster3) = cluster3$seurat_clusters
degs = FindAllMarkers(cluster3,only.pos = TRUE,return.thresh = 0.05,logfc.threshold = 0.25)
```

```r
degs[degs$cluster=='7',] %>% arrange(desc(avg_log2FC))
```

```r
cluster3_tpex_barcode = colnames(cluster3)[cluster3$seurat_clusters %in% c(0)]
cluster3_tint_barcode = colnames(cluster3)[cluster3$seurat_clusters %in% c(4)]
cluster3_tex_barcode = colnames(cluster3)[cluster3$seurat_clusters %in%  c(9,3,5)]
cluster3_teff1_barcode = colnames(cluster3)[cluster3$seurat_clusters %in% c(1)]
cluster3_teff2_barcode = colnames(cluster3)[cluster3$seurat_clusters %in% c(2)]
cluster3_teff3_barcode = colnames(cluster3)[cluster3$seurat_clusters %in% c(7,8)]
cluster3_teff4_barcode = colnames(cluster3)[cluster3$seurat_clusters %in% c(6)]
```

```r
cluster3$temp = 'temp'
cluster3$temp[colnames(cluster3) %in% cluster3_tpex_barcode] = 'CD8Tpex_TCF7'
cluster3$temp[colnames(cluster3) %in% cluster3_tint_barcode] = 'CD8Texint_TOX'
cluster3$temp[colnames(cluster3) %in% cluster3_tex_barcode] = 'CD8Tex_ENTPD1'
cluster3$temp[colnames(cluster3) %in% cluster3_teff1_barcode] = 'CD8Teff_CCL3'
cluster3$temp[colnames(cluster3) %in% cluster3_teff2_barcode] = 'CD8Teff_GNLY'
cluster3$temp[colnames(cluster3) %in% cluster3_teff3_barcode] = 'CD8Teff_GZMK' #这群细胞像是从Tpex向Teff过度的细胞，GZMB有一些增加，但是TOX依然有表达
cluster3$temp[colnames(cluster3) %in% cluster3_teff4_barcode] = 'CD8Teff_XCL1'
```

```r
options(repr.plot.width = 5.5, repr.plot.height = 4)
DimPlot(cluster3,group.by = 'temp',label = TRUE, repel = TRUE)
```

```r
options(repr.plot.width = 6, repr.plot.height = 4)
DotPlot(cluster3,features = c('TCF7','IL7R','XCL1','TOX','ENTPD1','GZMK','GNLY','CCL3'),group.by = 'temp') + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))
```

```r
options(repr.plot.width = 6, repr.plot.height = 5)
VlnPlot(cluster3,features = c('TCF7','IL7R','XCL1','TOX','ENTPD1','GZMK','GNLY','CCL3'),group.by = 'seurat_clusters', stack = TRUE, flip = TRUE)
```

```r
cd8t$curated[colnames(cd8t) %in% c(cluster3_tpex_barcode)] = 'CD8Tpex_TCF7'
cd8t$curated[colnames(cd8t) %in% c(cluster3_tint_barcode)] = 'CD8Texint_TOX'
cd8t$curated[colnames(cd8t) %in% c(cluster3_tex_barcode)] = 'CD8Tex_ENTPD1'
cd8t$curated[colnames(cd8t) %in% c(cluster3_teff1_barcode)] = 'CD8Teff_CCL3'
cd8t$curated[colnames(cd8t) %in% c(cluster3_teff2_barcode)] = 'CD8Teff_GNLY'
cd8t$curated[colnames(cd8t) %in% c(cluster3_teff3_barcode)] = 'CD8Teff_GZMK'
cd8t$curated[colnames(cd8t) %in% c(cluster3_teff4_barcode)] = 'CD8Teff_XCL1'
```
